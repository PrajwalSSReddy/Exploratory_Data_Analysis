```python
match_per_season=df.groupby(['session'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
session_data=df[['id','session']].merge(dfs,left_on='id',right_on='id',how='right').drop('id',axis=1)
session_data.head()
```

```python
session_dat=df[['id','session']].merge(dfs,left_on='id',right_on='id',how='right').drop('id',axis=1)
session_dat.head()
```

```python
dfs
```

```python
season=session_data.groupby(['session'])['total_runs'].sum().reset_index() # creating a variable and grouping by session and summing the totak runs by session and resetting the index vakue (which will reset the index value ordered after summing the total_runs)
p=season.set_index('session') #after above step arranging the index value according to the session after the summing the total runs 
ax=plt.axes()# it is used to make design like gray color and so on
ax.set(facecolor='gray') #setting the ackground color
sns.lineplot(data=p) #decalring the lineplot
plt.title('Total runs in each season',fontsize=12,fontweight='bold')
plt.show()
```

```python
season
```

```python
runs_per_match=pd.concat([match_per_season,season.iloc[:,1]],axis=1)
runs_per_match['Runs scored per match']=runs_per_match['total_runs']/runs_per_match['matches']
runs_per_match.set_index('session',inplace=True)
runs_per_match
```
